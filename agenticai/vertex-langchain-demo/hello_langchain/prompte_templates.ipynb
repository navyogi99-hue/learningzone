```python
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
project = os.getenv('GOOGLE_CLOUD_PROJECT')
llm = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash-lite",
    vertexai = True,
    project = project
)
```

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["country"],
    template="What is capital of the {country}"
)
```

```python
template.format(country = "India")
```

```python
'What is capital of the India'
```

```python
chain = template | llm 
response = chain.invoke({'country': 'India'})
```

```python
response.content
```

```python
template = PromptTemplate(
    input_variables=["country", "topic"],
    template="Explain about {topic} with respect to {country}"
)
```

```python
chain = template | llm
response = chain.invoke({"country": "India", "topic": "Independence"})
response.content
```

```python
response.pretty_print()
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage(content="You are math teacher for primary school"),
    HumanMessage(content="2 * 2"),
    AIMessage(content="This refers to multiplication \n Ensure you have memorized tables \n  2 * 2 = 4"),
    HumanMessage(content="5 + 5"),
    AIMessage(content="This refers to addition \n Count numbers on fingers \n 5 + 5 = 10"),
    HumanMessage(content="1+4")
]

response = llm.invoke(messages)
```

```python
response.pretty_print()
```
