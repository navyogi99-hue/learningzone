```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv
import os

```

```python
load_dotenv()
project = os.getenv('GOOGLE_CLOUD_PROJECT')
tavily_api_key = os.getenv('TAVILY_API_KEY')
```

```python
llm = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash-lite",
    vertexai = True,
    project = project
) 
```

```python
from langchain.tools import tool
@tool
def say_hello(name: str) -> str:
    """ This function is used to  greet the person


    Args:
        name(str): name

    Returns:
        str: greeting

    """
    return f"Hello {name}!"
```

```python
llm_with_tools = llm.bind_tools([say_hello])
```

```python
response =llm.invoke("greet shyam")
response
```

```python
response = llm_with_tools.invoke("Greet shyam")
response.pretty_print()
```

```python
from langchain.agents import create_agent
agent = create_agent(
    model=llm,
    tools=[say_hello]
)  
```

```python
agent_response =agent.invoke({
    "messages":[
         ("user","greet shyam")]
})
```

```python
agent_response['messages'][-1].pretty_print()
```

```python
agent_response['messages']
```
