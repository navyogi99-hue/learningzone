```python
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv
project=os.getenv('GGOGLE_CLOUD_PROJECT')
llm = ChatGoogleGenerativeAI(
    model= "gemini-2.5-flash-lite",
    vertexai=True,
    project=project
)
```

```python
template=PromptTemplate(
    input_variable=["country"],
    template="What is capital of the {country}"
)
```

```python
template.format(country="India")
```

```python
chain = template | llm
response=chain.invoke({"country":"India"})
response.content
```

```python
template=PromptTemplate(
    input_variables=["country","Topics"],
    template="Explian about {Topic} with respect to {country}"
)
```

```python
chain = template | llm
response = chain.invoke({"country":"India" ,"Topic":"1857 Revolution"})
response.content
```

```python
response.pretty_print()
```

```python
prompt=[
    SystemMessage(content="You are a primary teacher"),
    HumanMessage(content="2*5"),
    AIMessage(content="This refer to multiplication \n Ensure you have memorized tables \n 2*5=10"),
    HumanMessage(content="5+6"),
    AIMessage(content="This refer to addition \n count numbers on fingers \n 5+6=11"),
    HumanMessage(content="10-7")
]
response = llm.invoke(prompt)
```

```python
response.pretty_print()
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt= ChatPromptTemplate.from_messages([
    ("system","you are a primary teacher"),
    ("human","2*4"),
    ("ai","This refers to multiplication \n Ensure you have memorized tables \n 2*4=8"),
    ("human","5+7"),
    ("ai","This refers to addition \n Count numbers on figers \n 5+7=12"),
    ("human","{question}")
    
])
 
```

```python
chain = prompt | llm
response=chain.invoke({"question":"3*8"})
response.pretty_print()
```
