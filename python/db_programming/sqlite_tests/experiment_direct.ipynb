```python
import sqlite3
import os
```

```python
# create a connection
connection = sqlite3.connect("./data/student.db")
```

```python
# we need a cursor to execute SQL commands
cursor = connection.cursor()
```

```python
# using the cursor, we can execute SQL statements
CREATE_STUDENT_QUERY = """CREATE TABLE students (
    student_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    date_of_birth TEXT NOT NULL,
    gender TEXT,
    phone TEXT
    );
    """
cursor.execute(CREATE_STUDENT_QUERY)
connection.commit()
connection.close()
```

```python
# ideal way of doinng it
with sqlite3.connect("./data/student.db") as connection:
    cursor=connection.cursor()
    CREATE_STUDENT_QUERY = """CREATE TABLE students (
    student_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    date_of_birth TEXT,
    class_name TEXT NOT NULL,
    phone TEXT
);
"""
cursor.execute(CREATE_STUDENT_QUERY)
connection.commit()
connection.close()
```

```python
with sqlite3.connect("./data/student.db") as connection:
    cursor=connection.cursor()
    INSERT_STATIC_QUERRY = """INSERT INTO students (first_name, last_name, date_of_birth, gender, phone) VALUES ('Aarav', 'Sharma', '2010-05-12',  'male', '9876543210');"""
    cursor.execute(INSERT_STATIC_QUERRY)
    connection.commit()
```

```python
# Dynamic insertion
with sqlite3.connect("./data/student.db") as connection:
    cursor=connection.cursor()
    INSERT__QUERRY = """INSERT INTO students (first_name, last_name,date_of_birth, gender, phone) VALUES (?, ?, ?, ?, ?);"""

    VALUES = ('Rahul', 'Sharma', '2010-06-12', 'male', '9876543211')
    cursor.execute(INSERT__QUERRY, VALUES)
    connection.commit()
```
