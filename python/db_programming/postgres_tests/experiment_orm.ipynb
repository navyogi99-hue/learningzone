```python
from sqlalchemy.orm import DeclarativeBase
class Base(DeclarativeBase):
    pass
```

```python
from sqlalchemy import String, Integer,Float
from sqlalchemy import Column

class Course(Base):
    __tablename__ = "courses"
    id = Column(Integer, primary_key=True, index=True)
    name = Column(String(50))
    description = Column(String(200))
    price = Column(Float)
    
```

```python
Base.metadata.create_all(bind=engine)
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
```

```python
genai_course = Course(name="Generative AI", description="Learn about generative AI models and techniques.", price=199.99)
session.add(genai_course)
session.commit()
```

```python
python_course = Course(name="Python Programming", description="Learn Python programming from scratch.", price=99.99)
session.add(python_course)
session.commit()
```
